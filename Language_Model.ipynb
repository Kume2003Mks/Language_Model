‡∏Å‡∏≤‡∏£ clean data
---

```python
import numpy as np
import re

def clean_data(tweet):
    if type(tweet) == np.float64:
        return ""
    temp = re.sub("&\S+;", "", tweet)
    temp = re.sub("RT @[A-Za-z0-9_]+","", temp)
    temp = re.sub("@[A-Za-z0-9_]+","", temp)
    temp = re.sub("#[A-Za-z0-9_]+","", temp)
    temp = re.sub(r'http\S+', '', temp)
    temp = re.sub('[()!?]', ' ', temp)
    temp = re.sub('\[.*?\]',' ', temp)
    temp = re.sub("[^A-Za-z0-9]"," ", temp)
    return temp
```

```python
#import ngrams,bigrams and trigrams
import nltk
from nltk.util import ngrams
from nltk.util import bigrams
from nltk.util import trigrams
nltk.download('punkt')
from nltk import word_tokenize
from nltk.lm.preprocessing import padded_everygram_pipeline
from nltk.lm import MLE
```

---
Unigram ‡∏à‡∏≤‡∏Å‡∏Ñ‡∏≥‡∏£‡∏µ‡∏ß‡∏¥‡∏ß Positive
---

```python
pos_review = list(df[df['rating'] >= 4].clean_data.apply(word_tokenize))
ptrain, pvocab = padded_everygram_pipeline(1, pos_review)
pos_review
```

```python
lm = MLE(1)
lm.fit(ptrain, pvocab)
```

```python
for i in range(0,10):
  gen_text = lm.generate(15, text_seed=["<s>"])

  txt = ""
  for t in gen_text:
    if t != '</s>':
      txt += t + " "
    else:
      break
    
  print(txt)
  token_w = word_tokenize(txt)
  pperplex = list(ngrams(token_w,1))
  print("perplexity = %.2f\n"%lm.perplexity(pperplex))
  print("--------------------\n")
```

---
Bigram ‡∏à‡∏≤‡∏Å‡∏Ñ‡∏≥‡∏£‡∏µ‡∏ß‡∏¥‡∏ß Positive

```python
pos_review2 = list(df[df['rating'] >= 4].clean_data.apply(word_tokenize))
ptrain2, pvocab2 = padded_everygram_pipeline(2, pos_review2)
```

```python
lm2 = MLE(2)
lm2.fit(ptrain2, pvocab2)
```

```python
for i in range(0,10):
  gen_text = lm2.generate(15, text_seed=["<s>"])

  txt = ""
  for t in gen_text:
    if t != '</s>':
      txt += t + " "
    else:
      break
    
  print(txt)
  token_w = word_tokenize(txt)
  pperplex = list(bigrams(token_w))
  print("perplexity = %.2f\n"%lm2.perplexity(pperplex))
  print("--------------------\n")
```
